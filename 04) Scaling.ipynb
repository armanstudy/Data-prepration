## 01)Scaling

- Future scaling significantly improves the performance of some machine learning algorithms and does not work at all for others.

- **Scaling has 2 kinds**:
- a)**Normalize**:
- Normalization is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1.
- b)**Standardization**:
- it is another scaling method where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero, and the resultant distribution has a unit standard deviation. In this situation the outputs valuse will be betwwen -3 and 3,since:
$$\mu=1~~~~And~~~~\sigma=1$$

## 02)Normalize

- **a) Min-MAx-Normalize**

- Normalization transfers the data naumbers between 0 to 1.
- $$x_{new} = \frac{x-min}{max-min}$$


- min_max_normalization is sensetive to noise and outlier data:

- if the outier is a big number,then it causes all $x_{new}$ to becom near 0.

- Let's check them on python:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw  = pd.read_csv("heart.csv")
data = data_raw.copy()
```

```python
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler(feature_range=(0, 1))   #It is possible to change the range.
```

```python
mm = scale.fit_transform(data[["Age"]])
```

```python
print(mm)
```

```python
mm.min()
```

```python
mm.max()
```

```python
mm.std()
```

- Normalization is good to use when you know that the distribution of your data does not follow a Gaussian distribution.

**b)max_normalization**

- it is used in image data normaly which their values are positive.
- the value of each pixel will be divided by 255 in order to be normalized.
- In a dataframe all values in a colimn will be devided by the max number in that column.

```python
max_age = data.Age / data.Age.max()
```

```python
max_age.min()
```

```python
max_age.max()
```

```python
max_age.std()
```

## 03)Standardization

- we use standardization when our data does not follows normal distribution.

$$x_{}new = \frac{x-mean}{\sigma}$$

- This method is sensetive to outlier too.

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler()
```

```python
standard_age = standard.fit_transform(data[["Age"]])
```

```python
standard_age.min()
```

```python
standard_age.max()
```

```python
standard_age.std()
```

## 04)Binarizer

- Binarization transfer data to 0 and 1.
- we should define a threshold.
- the values less than and equal to threshold transfer to 0 and valuse more then that changes to 1.

```python
from sklearn.preprocessing import Binarizer
```

```python
binarize = Binarizer(threshold=40)
```

```python
bin_age = binarize.fit_transform(data[["Age"]])
```
