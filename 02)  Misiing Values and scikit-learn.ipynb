## 02)impor dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw  = pd.read_csv("G:\\Data_analyst\\05)data prepration\\heart.csv")
data = data_raw.copy()    #make a copy in order to compare data
```

## 03)Extarcting missing values

```python
data.isnull().sum()
```

- There is no null or NAN value in data,but in Cholesterol and RestingBP columns we have 0 values which are our missed values, since they are not allowed to be 0 in real world.

```python
data[data.Cholesterol == 0]
```

```python
data[data.RestingBP == 0]
```

- There is 172 zeros in cholestrol column and one 0 in RestingBP column and we should change them to null in order to be able to use them in SimpleImputer algorithm, since it does not know 0 as a missing value.

- We should replace 0 valuse here with null value, in order to be able to decide to replace the with a correct value. 

```python
data.replace({"Cholesterol":0},np.nan,inplace=True)
data.replace({"RestingBP":0},np.nan,inplace=True)
```

```python
data.isnull().sum()
```

## 04)Replacing missing valuse using Scikit-Learn.SimpleImputer

```python
from sklearn.impute import SimpleImputer
#Using simpleImputer we can fill missing values cells using mean, median, most_frequent(mode) and constant strategies.
```

__a)using mean__

```python
simple_imp = SimpleImputer(missing_values= np.nan , strategy="mean")  #making an object from simpleimputer
simple_imp.fit(data[["Cholesterol"]])                         #fit the model learn from "Cholesterol" and save it in simple_imp
data["Cholesterol"] = simple_imp.transform(data[["Cholesterol"]]) #transform implement the educated data       
```

##### Here our replacing missing values has been done.
- let's check the mean value and missing values.

- We can see that the mean in cholestrol column has been changed from 177 to 244

```python
data.describe()
```

- The mean value for Cholesterol has been calculated as 244.635389, that we calculated exactly this number using python code before.

```python
data.isnull().sum()
```

__b)using median strategy__

```python
#making a copy of our raw data
data2 = data_raw.copy()
data2.replace({"Cholesterol":0},np.nan,inplace=True)
simple_imp2 = SimpleImputer(missing_values= np.nan , strategy="median")
```

```python
data2["Cholesterol"] = simple_imp2.fit_transform(data2[["Cholesterol"]])
#here we have fitted and transformed data simeltaneously
```

```python
data2.describe()
```

__b)using median strategy__

```python
#making a copy of our raw data
data2 = data_raw.copy()
data2.replace({"Cholesterol":0},np.nan,inplace=True)
simple_imp2 = SimpleImputer(missing_values= np.nan , strategy="median")
```

```python
data2["Cholesterol"] = simple_imp2.fit_transform(data2[["Cholesterol"]])
#here we have fitted and transformed data simeltaneously
```

```python
data2.describe()
```

__c)using constant strategy__

```python
#making a copy of our raw data
data3 = data_raw.copy()
data3.replace({"Cholesterol":0},np.nan,inplace=True)
simple_imp3 = SimpleImputer(missing_values= np.nan , strategy="constant", fill_value=150)
```

```python
data3["Cholesterol"] = simple_imp3.fit_transform(data3[["Cholesterol"]])
#here we have fitted and transformed data simeltaneously
```

```python
data3.describe()
```
